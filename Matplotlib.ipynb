## 掌握Matplotlib基础绘图

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,1.1,0.1)

plt.figure()    #创建画布
plt.plot(x,x**2)#绘制图形
plt.plot(x,x**4)
plt.show()      #显示图形  
```

```python
x = np.arange(0,1.1,0.1)

plt.figure()    #创建画布
plt.plot(x,x**2)#绘制图形
plt.plot(x,x**4)
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('lines')
plt.legend(['y=x^2','y=x^4'])

plt.savefig('temp/lines.png')
plt.show()      #显示图形  
```

## 绘制散点图

```python
data = np.load('temp/2001-2019年劳动力与就业人员数据.npz',allow_pickle = True)
print(list(data.keys()))
columns = data['arr_0']
values = data['arr_1']
print(columns)
print(values)
values.shape
```

```python
plt.figure(figsize = (10,5))
plt.rcParams['font.sans-serif']='SimHei'#设置中文字体
plt.rcParams['axes.unicode_minus'] = False #解决在负号显示为方块的问题。


plt.scatter(values[:,0],values[:,1],marker = 'o',label='劳动力(万人)')
plt.scatter(values[:,0],values[:,2],marker = '*',label='就业人员(万)')
plt.scatter(values[:,0],values[:,3],marker = 'D',label='城镇就业人员(万人)')

plt.xticks(range(2001,2020,1),values[:,0],rotation = 45)
# plt.xticks(range(19),values[:,0],rotation=45)
plt.legend()
#plt.legend(handles=[line1, line2], labels=['Red Line', 'Blue Line'], loc='best')
#plt.legend(['劳动力(万人)','就业人员(万人)','城镇就业人员(万人)'])
# 在 `matplotlib` 中，`plt.legend()` 函数用来创建图例，它有多种使用方式。
# 当你传递一个列表给 `plt.legend()` 而不指定 `handles` 参数时，
# `matplotlib` 会尝试将列表中的标签与当前轴上通过 `label` 参数指定的最后绘制的元素进行匹配。
plt.title('2001-2019年就业人数统计散点图')
plt.ylabel('人数（万人）')
plt.savefig('temp/2001-2019年就业人数统计散点图')
plt.show()
```

## 绘制折线图

```python
plt.figure(figsize = (10,5))
plt.rcParams['font.sans-serif']='SimHei'#设置中文字体
plt.rcParams['axes.unicode_minus'] = False #解决在负号显示为方块的问题。


plt.plot(values[:,0],values[:,1],marker = 'o',label='劳动力(万人)',linestyle=':')
plt.plot(values[:,0],values[:,2],marker = '*',label='就业人员(万)')
plt.plot(values[:,0],values[:,3],marker = 'D',label='城镇就业人员(万人)')

plt.xticks(range(2001,2020,1),values[:,0],rotation = 45)
plt.title('2001-2019年就业人数统计折线图')
plt.ylabel('人数（万人）')
plt.legend()
plt.savefig('temp/2001-2019年就业人数统计折线图')
plt.show()
```

## 绘制直方图

```python
data = np.load('temp/2001-2019年劳动力与就业人员数据.npz',allow_pickle = True)
print(list(data.keys()))
columns = data['arr_0']
values = data['arr_1']
print(columns)
print(values)
values.shape
```

```python
values[18,1:17]
columns.dtype
```

```python
plt.figure(figsize = (10,5))
plt.rcParams['font.sans-serif']='SimHei'#设置中文字体
plt.rcParams['axes.unicode_minus'] = False #解决在负号显示为方块的问题。

plt.bar(columns[0,1:17],values[-1,1:17])
plt.xticks(range(0,16,1),columns[0,1:17],rotation = 45)
plt.title('2019年劳动力与就业人员统计直方图')
plt.ylabel('人数（万人）')
# plt.legend()

my_height = values[-1,1:17]
for i in range(len(my_height)):
    plt.text(i,my_height[i]+1000,my_height[i],va='bottom',ha='center' )
    print(i)
    
plt.savefig('temp/2019年劳动力与就业人员统计直方图')
plt.show()
```

## 绘制饼图

```python
plt.figure(figsize = (3,3))
plt.rcParams['font.sans-serif']='SimHei'#设置中文字体
plt.rcParams['axes.unicode_minus'] = False #解决在负号显示为方块的问题。

plt.pie(values[18,3:5],explode=[0.01,0],labels=['城镇就业人员(万人)','乡村就业人员(万人)'],autopct='%1.1f%%')

plt.savefig('temp/2019年劳动力与就业人员统计饼图')
plt.show()
```

## 绘制箱线图

```python
plt.figure(figsize = (3,3))
plt.rcParams['font.sans-serif']='SimHei'#设置中文字体
plt.rcParams['axes.unicode_minus'] = False #解决在负号显示为方块的问题。
labels=['城镇就业人员(万人)','乡村就业人员(万人)']
plt.boxplot(values[:,3:5],notch=True,labels=labels)
plt.xticks(range(1,3),labels,rotation = 45)
plt.savefig('temp/2019年劳动力与就业人员统计箱线图')
plt.show()
```
